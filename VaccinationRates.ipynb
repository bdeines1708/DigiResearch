```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
covid_data = pd.read_csv("covid-data.csv")

# Convert date to datetime format
covid_data['date'] = pd.to_datetime(covid_data['date'])

# Filter and prepare data for each section of analysis

# 1. Vaccinations and Total Cases Over Time
covid_daily = covid_data.groupby('date')[['total_cases', 'total_vaccinations']].sum()

plt.figure(figsize=(14, 6))
plt.plot(covid_daily.index, covid_daily['total_cases'], label='Total Cases', linewidth=2, color='blue')
plt.plot(covid_daily.index, covid_daily['total_vaccinations'], label='Total Vaccinations', linewidth=2, color='green')
plt.title('Global Total COVID-19 Cases and Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()

# 2. GDP and Poverty Levels vs. Total Cases
covid_economic_filtered = covid_data.dropna(subset=['total_cases', 'gdp_per_capita', 'extreme_poverty'])

fig, axs = plt.subplots(1, 2, figsize=(14, 6))
sns.scatterplot(x='gdp_per_capita', y='total_cases', data=covid_economic_filtered, ax=axs[0], color='purple')
axs[0].set_title('GDP per Capita vs. Total COVID-19 Cases')
axs[0].set_xlabel('GDP per Capita')
axs[0].set_ylabel('Total Cases')

sns.scatterplot(x='extreme_poverty', y='total_cases', data=covid_economic_filtered, ax=axs[1], color='orange')
axs[1].set_title('Extreme Poverty vs. Total COVID-19 Cases')
axs[1].set_xlabel('Extreme Poverty (%)')
axs[1].set_ylabel('Total Cases')
plt.tight_layout()
plt.show()

# 3. Age Demographics vs. Total Cases
covid_age_filtered = covid_data.dropna(subset=['total_cases', 'median_age', 'aged_65_older', 'aged_70_older'])

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
sns.scatterplot(x='median_age', y='total_cases', data=covid_age_filtered, ax=axs[0], color='cyan')
axs[0].set_title('Median Age vs. Total COVID-19 Cases')
axs[0].set_xlabel('Median Age')
axs[0].set_ylabel('Total Cases')

sns.scatterplot(x='aged_65_older', y='total_cases', data=covid_age_filtered, ax=axs[1], color='brown')
axs[1].set_title('Population Aged 65+ vs. Total COVID-19 Cases')
axs[1].set_xlabel('Population Aged 65+ (%)')
axs[1].set_ylabel('Total Cases')

sns.scatterplot(x='aged_70_older', y='total_cases', data=covid_age_filtered, ax=axs[2], color='green')
axs[2].set_title('Population Aged 70+ vs. Total COVID-19 Cases')
axs[2].set_xlabel('Population Aged 70+ (%)')
axs[2].set_ylabel('Total Cases')
plt.tight_layout()
plt.show()

# 4. Smoking Rates vs. Total Cases
covid_smoking_filtered = covid_data.dropna(subset=['total_cases', 'female_smokers', 'male_smokers'])

fig, axs = plt.subplots(1, 2, figsize=(14, 6))
sns.scatterplot(x='female_smokers', y='total_cases', data=covid_smoking_filtered, ax=axs[0], color='pink')
axs[0].set_title('Female Smokers vs. Total COVID-19 Cases')
axs[0].set_xlabel('Female Smokers (%)')
axs[0].set_ylabel('Total Cases')

sns.scatterplot(x='male_smokers', y='total_cases', data=covid_smoking_filtered, ax=axs[1], color='darkblue')
axs[1].set_title('Male Smokers vs. Total COVID-19 Cases')
axs[1].set_xlabel('Male Smokers (%)')
axs[1].set_ylabel('Total Cases')
plt.tight_layout()
plt.show()

# 5. Other Factors Potentially Affecting COVID-19 Cases
columns_of_interest = ['population_density', 'gdp_per_capita', 'extreme_poverty', 
                       'hospital_beds_per_thousand', 'life_expectancy', 'human_development_index', 'total_cases']
covid_factors_filtered = covid_data[columns_of_interest].dropna()

fig, axs = plt.subplots(2, 3, figsize=(18, 10))
sns.scatterplot(x='population_density', y='total_cases', data=covid_factors_filtered, ax=axs[0, 0], color='grey')
axs[0, 0].set_title('Population Density vs. Total Cases')
axs[0, 0].set_xlabel('Population Density')
axs[0, 0].set_ylabel('Total Cases')

sns.scatterplot(x='gdp_per_capita', y='total_cases', data=covid_factors_filtered, ax=axs[0, 1], color='purple')
axs[0, 1].set_title('GDP per Capita vs. Total Cases')
axs[0, 1].set_xlabel('GDP per Capita')
axs[0, 1].set_ylabel('Total Cases')

sns.scatterplot(x='extreme_poverty', y='total_cases', data=covid_factors_filtered, ax=axs[0, 2], color='orange')
axs[0, 2].set_title('Extreme Poverty vs. Total Cases')
axs[0, 2].set_xlabel('Extreme Poverty (%)')
axs[0, 2].set_ylabel('Total Cases')

sns.scatterplot(x='hospital_beds_per_thousand', y='total_cases', data=covid_factors_filtered, ax=axs[1, 0], color='blue')
axs[1, 0].set_title('Hospital Beds per Thousand vs. Total Cases')
axs[1, 0].set_xlabel('Hospital Beds per Thousand')
axs[1, 0].set_ylabel('Total Cases')

sns.scatterplot(x='life_expectancy', y='total_cases', data=covid_factors_filtered, ax=axs[1, 1], color='green')
axs[1, 1].set_title('Life Expectancy vs. Total Cases')
axs[1, 1].set_xlabel('Life Expectancy')
axs[1, 1].set_ylabel('Total Cases')

sns.scatterplot(x='human_development_index', y='total_cases', data=covid_factors_filtered, ax=axs[1, 2], color='red')
axs[1, 2].set_title('Human Development Index vs. Total Cases')
axs[1, 2].set_xlabel('Human Development Index')
axs[1, 2].set_ylabel('Total Cases')
plt.tight_layout()
plt.show()
```
